```python
# Importando pandas e numpy
import pandas as pd
import numpy as np
```

```python
# Crinado um dataframe(tabela) de exemplo
df = pd.DataFrame({
    'renda_media' :      [ 10, 12, 13, 21, 12, 11, 17],
    'bairro_classe' : [ 'A', 'B', 'C', 'B', 'B', 'A', 'A'],
    'cidade' :     [ 'Df', 'RJ', 'SP', 'RJ', 'SP', 'RJ', 'SP'],
    'sexo_maioria' : [ 'M', 'F', 'F', 'M', 'M', 'M', 'F'],
    'cor_favorita' : [ 'vermelha', np.NAN, 'amarela', np.NAN, 'preto', 'verde', 'vermelho']
})

   # print para ver a tabela
df
```

```python
# Selecionando todas as linhas que atendam a condição
df.loc[df.renda_media >= 15]
```

```python
# Selecionando pandas.DataFrame.LOC
df.loc[(df.renda_media >= 12) & (df.sexo_maioria == 'M')]
```

```python
# Fatiamento 'Query Builer'
df.loc[1:3]
```

```python
# Selecionando algumas colunas com operadores condicionais
df.loc[(df.renda_media >= 12), ['renda_media', 'cidade','sexo_maioria']]
```

```python
# Alterando (dando update) na colunas que segue a condicional
df.loc[(df.renda_media >= 20), ['bairro_classe']] = 'E'

```

```python
df
```

```python
# Alterando (update) várias colunas de acordo com a condicional
df.loc[(df.renda_media >= 12), ['bairro_classe', 'cidade']] = ['AA','CE']
        
```

```python
df
```

```python
# Selecionando linha pelo Índice
df.iloc[[0,2,6]]
```

```python
# Selecionando linhas com ídices e colunas espacíficas
df.iloc[[0,2],[1,3]]
```
